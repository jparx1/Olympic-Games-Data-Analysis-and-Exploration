# Olympic Games Data Analysis and Exploration

## Context:
* Data includes Olympic games information about athletes and medal results spanning from 1896 to 2016
* Original data source: https://www.kaggle.com/heesoo37/120-years-of-olympic-history-athletes-and-results

## Dataset Description:
* Will be working with athlete_events.csv file
* Each row corresponds to an individual athlete competing in an individual event.

## Objective:
* Clean and examine dataset
* Explore distributions of numerical and categorical features via statistic and plots (using Seaborn and pandas)
* Examine relationships of multiple features via statistics and plots (using Seaborn and pandas)

```python
#Import pandas and Seaborn
```

```python
import pandas as pd
import seaborn as sns
pd.option_context('mode.use_inf_as_na', True)
```

```python
#Import olympics dataset
```

```python
olympics=pd.read_csv('athlete_events.csv')
```

```python
#View info summary and head of df
```

```python
olympics.info()
olympics.head()
```

```python
#We have missing data, for purpose of exercise, let's impute based on Year, Age, Height, and Weight columns using iterative imputer
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
#Build list of columns for imputation
```

```python
cols_imp=['Year','Age','Height','Weight']
```

```python
#Create iterative imputer and set min and max value parameters to be that of the corresponding columns
```

```python
iter_imp=IterativeImputer(min_value=olympics[cols_imp].min(), max_value=olympics[cols_imp].max())
```

```python
#Apply imputer to fit and transform columns to an imputed array
```

```python
imputed_cols=iter_imp.fit_transform(olympics[cols_imp])
```

```python
#Assign back to df
```

```python
olympics[cols_imp]=imputed_cols
```

```python
#Fill in missing Medal values with 'NA'
```

```python
olympics['Medal']=olympics['Medal'].fillna('NA')
```

```python
#Calculate the above first and third quartiles for Age seen above along with IQR
```

```python
Q1_age=olympics['Age'].quantile(.25)
Q3_age=olympics['Age'].quantile(.75)
IQR_age=Q3_age-Q1_age

print(Q1_age)
print(Q3_age)
print(IQR_age)
```

```python
#Print out Age outlier thresholds based on IQR above
```

```python
upper_age_out=1.5*IQR_age+Q3_age
lower_age_out=Q1_age-1.5*IQR_age

print(upper_age_out)
print(lower_age_out)
```

```python
#Athletes younger than 13 and older than 37 may be considered outliers
```

```python
#What sports do young athletes play?
```

```python
low_age_out=(olympics['Age']<(Q1_age-1.5*IQR_age))
olympics.loc[low_age_out,'Sport']
```

```python
#Look at values
```

```python
olympics.loc[low_age_out,'Sport'].value_counts()
```

```python
#Younger athletes swim and figure skate
```

```python
#Do same for older athletes
```

```python
upp_age_out=(olympics['Age']>(Q3_age+1.5*IQR_age))
olympics.loc[upp_age_out,'Sport']
```

```python
#Look at values
```

```python
olympics.loc[upp_age_out,'Sport'].value_counts()
```
